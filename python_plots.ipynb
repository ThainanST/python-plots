## Generating figures with Python

Importing modules

```python
import matplotlib.pyplot as plt
import numpy as np
```

*Defining* the colors that can be used in the curves

```python
blu = [0.0000, 0.4470, 0.7410];
red = [0.8500, 0.3250, 0.0980];
yel = [0.9290, 0.6940, 0.1250];
mag = [0.4940, 0.1840, 0.5560];
gre = [0.4660, 0.6740, 0.1880];
cia = [0.3010, 0.7450, 0.9330];
vin = [0.6350, 0.0780, 0.1840];
```

Creating the data to be plotted

```python
f = 60;
w = 2*np.pi*f;
t = np.linspace(0, 2/f, 100, dtype=np.float32);

y1 = np.sin( w*t );
y2 = np.cos( w*t );
```

Plotting

```python
plt.figure(1, figsize=(7, 3));
line1 = plt.plot(t, y1, 'k--', label='sin(wt)');
line2 = plt.plot(t, y2, 'k:', label='cos(wt)');

plt.setp(line1, 'color', red, 'linewidth', 2.0, 'linestyle', '-');
plt.setp(line2, 'color', blu, 'linewidth', 1.0, 'linestyle', '-');

plt.legend(loc='upper right', shadow=True, fontsize='x-large');
plt.title('My title');
plt.xlabel("Time, s");
plt.ylabel("Voltage, V");
plt.grid();
plt.axis([0, 2/f, -1, 1]); # [xmin, xmax, ymin, ymax]

nameFIG = "figure001.pdf"
plt.savefig(nameFIG)
```

Using subplots

```python
plt.figure(2, figsize=(10, 3));

plt.subplot(121)
lines = plt.plot(t, y1)
plt.setp(lines, 'color', red, 'linewidth', 2.0)
plt.axis([0, 2/f, -1, 1]); # [xmin, xmax, ymin, ymax]
plt.title('My Title 1') 
plt.xlabel("Time, s");
plt.ylabel("Voltage, V");
plt.grid()

plt.subplot(122)
lines = plt.plot(t, y2)
plt.setp(lines, 'color', blu, 'linewidth', 2.0)
plt.axis([0, 2/f, -1, 1]); # [xmin, xmax, ymin, ymax]
plt.title('My Title 2') 
plt.xlabel("Time, s");
plt.ylabel("Current, V");
plt.tight_layout(pad=1.0)
plt.grid()

nameFIG = "figure002.pdf"
plt.savefig(nameFIG)
```
